```python
import pandas as pd
import numpy as np
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

**Input the csv file**

```python
data1= pd.read_csv("PD 2022 Wk 1 Input - Input2.csv")
data1.head(10)
```

```python
data2= pd.read_csv("PD 2022 WK 3 Grades.csv")
data2.head(10)
```

**Join the data sets together to give us the grades per student**

```python
data=pd.merge(data1,data2, left_on='id', right_on='Student ID')
```

**Remove the parental data fields, they aren't needed for the challenge this week**

```python
data = data.drop(columns=["Parental Contact Name_1", "Parental Contact Name_2", "Preferred Contact Employer", "Parental Contact"])
```

**Pivot the data to create one row of data per student and subject <br>
Rename the pivoted fields to Subject and Score**

```python
data = pd.melt(data, id_vars=['Student ID','pupil first name','pupil last name','gender','Date of Birth'], value_vars=["Maths","English","Spanish","Science","Art","History","Geography"],
                   var_name="Subject", value_name='Score')
```

**Create an average score per student based on all of their grades**

```python
data["Student's Avg Score"] = data.groupby(['Student ID'])['Score'].transform('mean')
```

**Create a field that records whether the student passed each subject <br>
Pass mark is 75 and above in all subjects**

```python
data.loc[data['Score'] >=75, 'If_Passed'] = 'Passed'  
data.loc[data['Score'] <75, 'If_Passed'] = 'Failed'
```

**Aggregate the data per student to count how many subjects each student passed**

```python
df=pd.DataFrame(data.groupby(['Student ID','gender',"Student's Avg Score"]).apply(lambda x: x[x['If_Passed'] == 'Passed' ]['If_Passed'].count()))
df=df.reset_index()
df.columns=['Student ID','gender',"Student's Avg Score","Passed Subjects"]
df
```

**Round the average score per student to one decimal place**

```python
df["Student's Avg Score"]=df["Student's Avg Score"].round(decimals = 1)
```

**Remove any unnecessary columns of data**

**Output the data**

```python
output_columns =["Passed Subjects","Student's Avg Score",'Student ID','gender']
df = df[output_columns]


#Output the data 
df.to_csv('PD 2022 Week 3 Output.csv', index=False)

```
